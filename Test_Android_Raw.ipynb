```python
def ensemble_classifier(pressure_features, acc_features):
    x = pressure_features.values[:,list(range(9))]
    y = pressure_features.values[:,[9]]
    ypred = []
    for row in x:
        # Vertical Transition Detection
        out_vt = int(vert_clf.predict(row[:-2].reshape(1,-1)))
        if out_vt == 0:
            out = 0
        else:
            # Direction
#             out_dir = 0
            out_dir = int(dir_clf.predict(row[5].reshape(-1,1)))
            # Elevator Detection
            out_ele = int(ele_clf. predict(row[:-2].reshape(1,-1)))
            if out_ele == 0:
                # Elevator
                if out_dir == 0:
                    # Up
                    out = 5
                else:
                    # Down
                    out = 6
            else:
                acc_rows = acc_features[(acc_features['ts_min'] >= row[-1]) & (acc_features['ts_min'] <= row[-2])]
#                 acc_rows = acc_features[((acc_features['ts_min'] >= 1519395639221) & (acc_features['ts_min'] <= 1519395644081)) |
#                        ((acc_features['ts_max'] >= 1519395639221) & (acc_features['ts_max'] <= 1519395644081))]
                acc_out = {}
                if len(acc_rows) < 1:
#                     print(acc_rows)
#                     print(int(row[-1]), int(row[-2]))
#                     input("no rows")
                    out = -1
                    ypred.append(out)
                    continue
                for arow in acc_rows.values[:,[2,4,9,12,16,18,19]]:
                    ao = int(esc_stair_clf.predict(arow.reshape(1,-1)))
#                     print(ao)
                    try:
                        acc_out[int(ao)] += 1
                    except KeyError:
                        acc_out[int(ao)] = 1
                if len(acc_out.values()) != 1:
                    ao = max(acc_out.items(), key = lambda x: x[1])[0]
                    print(ao, acc_out.items())
                # Stairs
                if ao == 0:
#                     print(ao)
                    if out_dir == 0:
                        out = 1
                    else:
                        out = 2
                # Escalator
                else:
#                     print(ao)
                    if out_dir == 0:
                        out = 3
                    else:
                        out = 4
#                 else:
#                     max(stats.items(), key = lambda x: x[1])[0]
#                     print(acc_out)
#                     out = -2
                
        ypred.append(out)
    return y, ypred
    
```

```python
pressure_features = pd.concat([w_frame, su_frame, sd_frame, lu_frame, ld_frame, eu_frame, ed_frame])
print(len(pressure_features))
pressure_features = pressure_features.reset_index(drop=True)
```

```python
acc_features = pd.concat([sua_frame, sda_frame, eua_frame, eda_frame, lua_frame, lda_frame, wa_frame])
print(len(acc_features))
acc_features = acc_features.reset_index(drop=True)
```

```python
y, ypred = ensemble_classifier(pressure_features, acc_features)
```
