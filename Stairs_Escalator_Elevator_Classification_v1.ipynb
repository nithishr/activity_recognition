## Stairs vs Escalator + Elevator

```python
w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame = create_features_from_files(2000, 4000)
print_characteristics(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame)
```

```python
def create_dataset_vertical_1(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame):
    s_frame = pd.concat([su_frame, sd_frame])
    s_frame['label_v1'] = 0
    o_frame = pd.concat([eu_frame, ed_frame, lu_frame, ld_frame])
    o_frame['label_v1'] = 1
    return s_frame, o_frame
```

```python
s_frame, m_frame = create_dataset_vertical_1(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame)
display(s_frame, m_frame)
```

```python
def visualize_vertical_v1_features(s_frame, m_frame):
    s_features_array = s_frame.as_matrix(columns=s_frame.columns)
    m_features_array = m_frame.as_matrix(columns=m_frame.columns)
    X = np.concatenate([s_features_array, m_features_array])
    print(X.shape)
    
    windows_map = { 0:{'values':X[:,0], 'legend': 'skewness'}, 1:{'values':X[:,5], 'legend': 'gradient'},
               2:{'values':X[:,3], 'legend':'kurtosis'}, 3:{'values':X[:,4], 'legend':'std deviation'},
               4:{'values':X[:,1], 'legend':'percentile_windows'}, 5:{'values':X[:,2], 'legend': 'iqr'},
               6:{'values':X[:,6], 'legend':'norm'}, 7:{'values':X[:,8], 'legend':'labels'}}

    combos = list(combinations(list(range(len(windows_map.keys())-1)), 2))
#     print(len(combos))

    for combo in combos:
        fig = plt.figure()
        ax1 = fig.add_subplot(111)
        ax1.scatter(windows_map[combo[0]]['values'][:len(s_frame)], windows_map[combo[1]]['values'][:len(s_frame)], c='r', alpha=0.5, label='stairs')
        ax1.scatter(windows_map[combo[0]]['values'][len(s_frame):], windows_map[combo[1]]['values'][len(s_frame):], c='b', alpha=0.5, label='others')
        ax = plt.subplot()
        ax.set_xlabel(windows_map[combo[0]]['legend'])
        ax.set_ylabel(windows_map[combo[1]]['legend'])
        ax.legend()
        ax.grid(True)
        plt.show()
```

```python
visualize_vertical_v1_features(s_frame,m_frame)
```

```python
def create_vertical_v1_dataset(s_frame, m_frame):
    s_features_array = s_frame.as_matrix(columns=s_frame.columns)
    m_features_array = m_frame.as_matrix(columns=m_frame.columns)
    X = np.concatenate([s_features_array, m_features_array])
    Y = X[:,8]
    X = X[:,:7]
    return X, Y
```

```python
X,Y = create_vertical_v1_dataset(s_frame, m_frame)
print(X.shape, Y.shape)
```

```python
avg_acc, avg_f, l_list = evaluate_stratified_ml_algorithms(X, Y)
print(avg_acc, avg_f)
```

```python
plot_performance(avg_acc, avg_f, l_list)
```

```python
np.average(Y)
```

## Escalator vs Elevator

```python
def create_dataset_esc_ele(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame):
    e_frame = pd.concat([eu_frame, ed_frame])
    e_frame['label_v11'] = 0
    l_frame = pd.concat([lu_frame, ld_frame])
    l_frame['label_v11'] = 1
    return e_frame, l_frame
```

```python
def create_esc_ele_dataset(e_frame, l_frame):
    e_features_array = e_frame.as_matrix(columns=e_frame.columns)
    l_features_array = l_frame.as_matrix(columns=l_frame.columns)
    X = np.concatenate([e_features_array, l_features_array])
    Y = X[:,8]
    X = X[:,:7]
    return X, Y
```

```python
def visualize_esc_ele_features(e_frame, l_frame):
    e_features_array = e_frame.as_matrix(columns=e_frame.columns)
    l_features_array = l_frame.as_matrix(columns=l_frame.columns)
    X = np.concatenate([e_features_array, l_features_array])
    print(X.shape)
    
    windows_map = { 0:{'values':X[:,0], 'legend': 'skewness'}, 1:{'values':X[:,5], 'legend': 'gradient'},
               2:{'values':X[:,3], 'legend':'kurtosis'}, 3:{'values':X[:,4], 'legend':'std deviation'},
               4:{'values':X[:,1], 'legend':'percentile_windows'}, 5:{'values':X[:,2], 'legend': 'iqr'},
               6:{'values':X[:,6], 'legend':'norm'}, 7:{'values':X[:,8], 'legend':'labels'}}

    combos = list(combinations(list(range(len(windows_map.keys())-1)), 2))
#     print(len(combos))

    for combo in combos:
        fig = plt.figure()
        ax1 = fig.add_subplot(111)
        ax1.scatter(windows_map[combo[0]]['values'][:len(e_frame)], windows_map[combo[1]]['values'][:len(e_frame)], c='r', alpha=0.5, label='escalator')
        ax1.scatter(windows_map[combo[0]]['values'][len(e_frame):], windows_map[combo[1]]['values'][len(e_frame):], c='b', alpha=0.5, label='elevator')
        ax = plt.subplot()
        ax.set_xlabel(windows_map[combo[0]]['legend'])
        ax.set_ylabel(windows_map[combo[1]]['legend'])
        ax.legend()
        ax.grid(True)
        plt.show()
```

```python
w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame = create_features_from_files(2000, 4000)
print_characteristics(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame)
e_frame, l_frame = create_dataset_esc_ele(w_frame, su_frame, sd_frame, eu_frame, ed_frame, lu_frame, ld_frame)
display(e_frame, l_frame)
```

```python
visualize_esc_ele_features(e_frame,l_frame)
```

```python
X,Y = create_esc_ele_dataset(e_frame, l_frame)
print(X.shape, Y.shape)
avg_acc, avg_f, l_list = evaluate_stratified_ml_algorithms(X, Y)
print(avg_acc, avg_f)
```
