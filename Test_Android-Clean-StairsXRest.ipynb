```python
def ensemble_classifier(pressure_features, acc_features):
    x = pressure_features.values[:,list(range(9))]
    y = pressure_features.values[:,[9]]
    ypred = []
    for row in x:
        # Vertical Transition Detection
        out_vt = int(vert_clf.predict(row[:-2].reshape(1,-1)))
        if out_vt == 0:
            out = 0
        else:
            # Direction
#             out_dir = 0
            out_dir = int(dir_clf.predict(row[5].reshape(-1,1)))
            # Stairs Detection
            acc_rows = acc_features[((acc_features['ts_min'] >= row[-1]) & (acc_features['ts_min'] <= row[-2])) |
                       ((acc_features['ts_max'] >= row[-1]) & (acc_features['ts_max'] <= row[-2]))]
            acc_out = {}
            if len(acc_rows) < 1:
#                     print(acc_rows)
#                     print(int(row[-1]), int(row[-2]))
#                     input("no rows")
                out = -1
                ypred.append(out)
                continue
            for arow in acc_rows.values[:,list(range(21))]:
                ao = int(stair_rest_clf.predict(arow.reshape(1,-1)))
#                     print(ao)
                try:
                    acc_out[int(ao)] += 1
                except KeyError:
                    acc_out[int(ao)] = 1
            if len(acc_out.values()) != 1:
                ao = max(acc_out.items(), key = lambda x: x[1])[0]
                print(ao, acc_out.items())
#             out_sta = int(sta_clf. predict(row[:-2].reshape(1,-1)))
            if ao == 0:
                # Stairs
                if out_dir == 0:
                    # Up
                    out = 1
                else:
                    # Down
                    out = 2
            else:
                out_esc_ele = int(clf_esc_ele.predict(row[:-2].reshape(1,-1)))
                # Elevator
                if out_esc_ele == 1:
#                     print(ao)
                    if out_dir == 0:
                        out = 1
                    else:
                        out = 2
                # Escalator
                else:
#                     print(ao)
                    if out_dir == 0:
                        out = 3
                    else:
                        out = 4
        ypred.append(out)
    return y, ypred
    
```
